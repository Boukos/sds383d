```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#non-constant variance
N=100
x = np.ones((N,2))
x[:,1] = np.random.normal(size=N)
beta = np.array([-1.,1.])
sigma_beta = np.array([-1.,1.])
sigma = np.exp(np.dot(x,sigma_beta))
f = np.dot(x,beta)
y = np.random.normal(f,sigma)

plt.plot(x[:,1],y,'x')
plt.show()
```

```python
#multimodal
N=200
x = np.ones((N,2))
x[:,1] = np.random.normal(size=N)
pie = 0.6
beta0 = np.array([-1.,-0.2])
beta1 = np.array([-0.5,-4.])
y = np.zeros(N)
for n in range(N):
    r = np.random.rand()
    if r<pie:
        y[n] = np.random.normal(np.dot(x[n,:],beta0),)
    else:
        y[n] = np.random.normal(np.dot(x[n,:],beta1),)
        
plt.plot(x[:,1],y,'x')
plt.show()
```

```python
pie = 0.4
mu0 = np.array([-1.,-2.])
mu1 = np.array([2.,2.])
N=500
X = np.zeros((N,2))
Z = np.zeros(N)
for i in range(N):
    r = np.random.rand()
    if r<pie:
        X[i,:] = np.random.normal(mu0)
    else:
        X[i,:] = np.random.normal(mu1)
        Z[i] = 1
plt.plot(X[:,0],X[:,1],'x')
plt.axis([-7,7,-7,7])
plt.show()
plt.plot(X[np.where(Z==1),0],X[np.where(Z==1),1],'ko')
plt.plot(X[np.where(Z==0),0],X[np.where(Z==0),1],'rx')

plt.axis([-7,7,-7,7])
plt.show()
```

```python
pie = 0.4
mu0 = np.array([-1.,-2.])
mu1 = np.array([2.,2.])
N=500
X = np.zeros((N,2))
Z = np.zeros(N)
for i in range(N):
    r = np.random.rand()
    prec = np.random.gamma(1.,1.)
    sigma = 1./np.sqrt(prec)
    if r<pie:
        X[i,:] = np.random.normal(mu0,prec)
    else:
        X[i,:] = np.random.normal(mu1,prec)
        Z[i] = 1
plt.plot(X[:,0],X[:,1],'x')
plt.axis([-7,7,-7,7])
plt.show()

plt.plot(X[np.where(Z==0),0],X[np.where(Z==0),1],'rx')
plt.plot(X[np.where(Z==1),0],X[np.where(Z==1),1],'ko')
plt.axis([-7,7,-7,7])
plt.show()
```
